```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import json
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
```

First I will merge my 3 `.json files` (`StreamingHistory_music_0.json`, `StreamingHistory_music_1.json`, `StreamingHistory_music_2.json`.) into 1 because they contain the same data.

```python
import os

# Get current working directory
current_directory = os.getcwd()

# Initialize an empty list to store data from all JSON files
merged_data = []

# Loop through each file in the current directory
for filename in os.listdir(current_directory):
    if filename.startswith('StreamingHistory_music_') and filename.endswith('.json'):
        with open(os.path.join(current_directory, filename)) as f:
            # Load JSON data from the file
            data = json.load(f)
            # Append data to the list
            merged_data.extend(data)

# Specify the new filename
new_filename = 'StreamingHistory.json'

# Write the combined data to the new JSON file
with open(new_filename, 'w') as f:
    json.dump(merged_data, f)

print("Merged JSON files successfully. Combined data saved as '{}'.".format(new_filename))
```

In the context of the script we've been working on, `os.path.join()` **is used to construct file paths**, `os.path.exists()` **is used to check if a file exists**, and `os.getcwd()` **is used to get the current working directory.**

Second, I will **convert** the `.json` files to `.csv` because I find it much easier to do so since I am used to it working with that file format up until now.

```python
import csv


# Get the current working directory
current_directory = os.getcwd()

# List of JSON files to convert
json_files = ['Playlist1.json', 'StreamingHistory.json', 'YourLibrary.json']

# Loop through each JSON file
for filename in json_files:
    # Construct the full file path
    file_path = os.path.join(current_directory, filename)
    
    # Check if the file exists
    if os.path.exists(file_path):
        try:
            # Open JSON file and load data
            with open(file_path) as f:
                data = json.load(f)
            
            # Specify CSV filename
            csv_filename = os.path.splitext(filename)[0] + '.csv'

            # Open CSV file for writing
            with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
                # Create a CSV writer object
                writer = csv.writer(csvfile)
                
                # Write headers (assuming keys of the first dictionary as headers)
                writer.writerow(data[0].keys())
                
                # Write data rows
                for item in data:
                    writer.writerow(item.values())

            print(f"Converted '{filename}' to '{csv_filename}'.")
        except Exception as e:
            print(f"Error processing '{filename}': {str(e)}")
    else:
        print(f"File '{filename}' not found in directory.")

print("Conversion completed.")
```

The issues encountered while converting `Playlist1.jso`n and `YourLibrary.json` were due to differences in their structures compared to the assumptions made by the initial script. Here's a breakdown of the issues and their resolutions:

- `Playlist1.json`: Initially, the script assumed that the JSON file directly contained playlists with their respective tracks. However, the actual structure included a "playlists" key containing a list of playlists, each with its own tracks. This caused a `KeyError` because the script attempted to access keys at the top level that didn't exist. The resolution involved iterating over each playlist within the "playlists" key and then extracting track information accordingly.

```python
current_directory = os.getcwd()


json_file = 'Playlist1.json'

# Construct the full file path
file_path = os.path.join(current_directory, json_file)

# Check if the file exists
if os.path.exists(file_path):
    try:
        # Open JSON file and load data
        with open(file_path) as f:
            data = json.load(f)
        
        # Check if the JSON data contains playlists
        if 'playlists' in data:
            # Specify CSV filename
            csv_filename = os.path.splitext(json_file)[0] + '.csv'
            
            # Open CSV file for writing
            with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
                # Create a CSV writer object
                writer = csv.writer(csvfile)
                
                # Write headers
                writer.writerow(['Playlist Name', 'Track Name', 'Artist Name', 'Album Name', 'Track URI', 'Added Date'])
                
                # Iterate over each playlist
                for playlist in data['playlists']:
                    playlist_name = playlist['name']
                    # Write playlist name for each track
                    for item in playlist['items']:
                        track = item['track']
                        writer.writerow([playlist_name, track['trackName'], track['artistName'], track['albumName'], track['trackUri'], item['addedDate']])
            
            print(f"Converted all playlists in '{json_file}' to '{csv_filename}'.")
        else:
            print(f"No 'playlists' found in '{json_file}'. Skipping conversion.")
    except Exception as e:
        print(f"Error processing '{json_file}': {type(e).__name__} - {str(e)}")
else:
    print(f"File '{json_file}' not found in directory.")
```

- `YourLibrary.jso`n: Similar to `Playlist1.json`, the initial assumption was that the JSON file directly contained track information. However, the actual structure included a "tracks" key containing a list of tracks. Again, the script encountered a `KeyError` because it tried to access keys that didn't exist at the top level. The solution involved iterating over each track within the "tracks" key and extracting the necessary information.

```python
current_directory = os.getcwd()


json_file = 'YourLibrary.json'


file_path = os.path.join(current_directory, json_file)


if os.path.exists(file_path):
    try:
        # Open JSON file and load data
        with open(file_path) as f:
            data = json.load(f)
        
        # Check if the JSON data contains the expected structure
        if 'tracks' in data:
            # Specify CSV filename
            csv_filename = os.path.splitext(json_file)[0] + '.csv'
            
            # Open CSV file for writing
            with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
                # Create a CSV writer object
                writer = csv.writer(csvfile)
                
                # Write headers
                writer.writerow(['Artist', 'Album', 'Track Name', 'Track URI'])
                
                # Write data rows
                for track in data['tracks']:
                    writer.writerow([track['artist'], track['album'], track['track'], track['uri']])
            
            print(f"Converted 'YourLibrary.json' to '{csv_filename}'.")
        else:
            print(f"No 'tracks' found in '{json_file}'. Skipping conversion.")
    except Exception as e:
        print(f"Error processing '{json_file}': {type(e).__name__} - {str(e)}")
else:
    print(f"File '{json_file}' not found in directory.")
```

By adjusting the script to correctly handle the structures of `Playlist1.json` and `YourLibrary.json`, we were able to successfully convert them into `.csv` files.
**Understanding the structure of the JSON data and modifying the script accordingly is crucial for accurate data processing.**

# Or lets do the same action on streaming_history so that we also find the most streamed ones.

```python
# Filter tracks with exact string matches
exact_string_matches = streaming_history[streaming_history['trackName'].str.lower() == 'candy']

# Group by track name and sum the msPlayed to find total streams
stream_count = exact_string_matches.groupby(['artistName', 'trackName'])['msPlayed'].sum()

# Merge tracks with the same name into one track and sum their msPlayed
merged_tracks = stream_count.groupby(level=[0, 1]).sum()

# Sort the merged tracks by total streams in descending order
sorted_merged_tracks = merged_tracks.sort_values(ascending=False)

# Display the top 5 merged tracks
print("Top 5 merged tracks with exact string match 'Candy':")
for i, ((artist_name, track_name), total_streams) in enumerate(sorted_merged_tracks.items()):
    if i >= 5:
        break
    print(f"{artist_name} - {track_name}: {total_streams} msPlayed")
```

## Most featured Artist in all my playlists?

```python
from collections import Counter

# Create a Counter to store the count of each artist
artist_counter = Counter()

# Iterate over each playlist
for playlist_name, group in playlist1.groupby('Playlist Name'):
    # Concatenate all track artists in the playlist and update the Counter
    artist_counter.update(group['Artist Name'])

# Find the most featured artist
most_featured_artist = artist_counter.most_common(1)[0][0]
num_appearances = artist_counter.most_common(1)[0][1]

# Print the result
print(f"The most featured artist in all your playlists is '{most_featured_artist}' with {num_appearances} appearances.")
```

## Most featured Album in all my playlists?

```python
from collections import Counter

# Create a Counter to store the count of each album
album_counter = Counter()

# Iterate over each playlist
for playlist_name, group in playlist1.groupby('Playlist Name'):
    # Concatenate all track albums in the playlist and update the Counter
    album_counter.update(group['Album Name'])

# Find the top 5 most featured albums
top_5_albums = album_counter.most_common(5)

# Print the result
print("The top 5 most featured albums in all your playlists are:")
for album, num_appearances in top_5_albums:
    print(f"{album}: {num_appearances} appearances")
```

## Analyzing trends in playlist creation over time (e.g., frequency of playlist additions).


```python


# Convert 'Added Date' column to datetime
playlist1['Added Date'] = pd.to_datetime(playlist1['Added Date'])

# Extract date and month from 'Added Date'
playlist1['Added Month'] = playlist1['Added Date'].dt.to_period('M')

# Group by 'Added Month' and count the number of playlist additions
playlist_counts = playlist1.groupby('Added Month').size()

# Ensure all months are included in the index
all_months = pd.period_range(start=playlist_counts.index.min(), end=playlist_counts.index.max(), freq='M')
playlist_counts = playlist_counts.reindex(all_months, fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))
playlist_counts.plot(marker='o', linestyle='-', color='b')

# Formatting
plt.title('Playlist Additions Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Playlist Additions')
plt.xticks(rotation=45)
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

## How much of my time have I spent listening to music on Spotify?

```python
# Calculate total duration in milliseconds
total_ms_played = streaming_history['msPlayed'].sum()

# Convert milliseconds to days, hours, minutes, and seconds
total_seconds = total_ms_played // 1000
days = total_seconds // (24 * 3600)
hours = (total_seconds % (24 * 3600)) // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

# Print the result
print(f"Total listening time on Spotify: {days} days, {hours} hours, {minutes} minutes, and {seconds} seconds.")
```

# Average streaming time/date?

```python

# Extract date from 'endTime'
streaming_history['date'] = streaming_history['endTime'].dt.date

# Convert 'msPlayed' to seconds
streaming_history['seconds_played'] = streaming_history['msPlayed'] / 1000  # Convert milliseconds to seconds

# Group by date and sum 'seconds_played' for each day
daily_seconds_played = streaming_history.groupby('date')['seconds_played'].sum()

# Calculate the total streaming time in seconds
total_seconds_played = daily_seconds_played.sum()

# Calculate the total number of days
num_days = len(daily_seconds_played)

# Calculate the average streaming time per day in seconds
average_seconds_per_day = total_seconds_played / num_days

# Convert seconds to hours
average_hours_per_day = average_seconds_per_day / 3600  # 1 hour = 3600 seconds

# Print the result
print(f"Average streaming time per day: {average_hours_per_day:.2f} hours")
```

# To determine the most popular artists based on their presence in playlists and listening history, I will follow these steps:

```python
# Load the playlists data
playlist1 = pd.read_csv('Playlist1.csv')

# Count occurrences of each artist and album in playlists
artist_playlist_counts = playlist1['Artist Name'].value_counts()

# Determine most frequent artists and albums added to playlists
most_frequent_artists = artist_playlist_counts.head(10)

# Display results
print("Most frequent artists in playlists:")
print(most_frequent_artists)
```

```python
# Load the streaming history data
streaming_history = pd.read_csv('StreamingHistory.csv')

# Count occurrences of each artist in the streaming history
artist_stream_counts = streaming_history['artistName'].value_counts()


# Determine most frequently streamed artists
most_streamed_artists = artist_stream_counts.head(10)

# Display results
print("Most frequently streamed artists:")
print(most_streamed_artists)
```

```python
# Calculate the counts of each artist in the streaming history
artist_counts = streaming_history['artistName'].value_counts()

# Normalize the counts by dividing by the total count and multiplying by 100 to get percentages
normalized_artist_counts = artist_counts / artist_counts.sum() * 100

# Filter out less frequent occurrences (optional)
threshold = 2  # Adjust as needed
filtered_artist_counts = normalized_artist_counts[normalized_artist_counts >= threshold]

# Plotting the most popular artists
plt.figure(figsize=(10, 6))
filtered_artist_counts.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Most Popular Artists in Streaming History')
plt.xlabel('Artist')
plt.ylabel('Percentage of Total Listening Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Correlation between playlist additions and artist popularity

```python
# Step 1: Calculate the number of playlist additions for each artist
playlist_additions_by_artist = playlist1.groupby('Artist Name').size()

# Step 2: Determine the popularity of each artist based on their presence in playlists
# You can use the number of playlist additions or the proportion of playlist additions among all artists as a measure of popularity

# Calculate the total number of playlist additions
total_playlist_additions = playlist_additions_by_artist.sum()

# Calculate the proportion of playlist additions for each artist
artist_popularity = playlist_additions_by_artist / total_playlist_additions

# Step 3: Analyze the correlation between playlist additions and artist popularity
# You can use correlation coefficients such as Pearson's correlation coefficient or Spearman's rank correlation coefficient

# For example, using Pearson's correlation coefficient:
correlation = playlist_additions_by_artist.corr(artist_popularity)

# Print the correlation coefficient
print("Correlation between playlist additions and artist popularity:", correlation)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(playlist_additions_by_artist.values, artist_popularity.values, color='b', alpha=0.5)

# Formatting
plt.title('Playlist Additions vs. Artist Popularity')
plt.xlabel('Number of Playlist Additions')
plt.ylabel('Artist Popularity (Proportion)')
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

If the scatter plot appears as a straight diagonal line, it indicates a strong positive correlation between playlist additions and artist popularity. In other words, as the number of playlist additions for an artist increases, their overall popularity also tends to increase. This suggests that artists whose songs are frequently added to playlists are generally more popular among listeners.

This correlation makes intuitive sense because popular artists are more likely to have their songs added to playlists by a larger number of listeners. Conversely, less popular artists may have fewer playlist additions.

So, seeing a straight diagonal line in the scatter plot is a positive indicator of the relationship between playlist additions and artist popularity.

```python
# Get the top streamed track and its corresponding artist
most_streamed_track = top_tracks.iloc[0]

# Get the total streaming time for the most streamed track
most_streamed_time = most_streamed_track['timeStreaming'].total_seconds()

# Calculate total streaming time for each artist
artist_total_streaming_time = streaming_history.groupby('artistName')['timeStreaming'].sum().astype('timedelta64[s]').to_dict()

# Get the top 10 artists
top_artists = sorted(artist_total_streaming_time, key=artist_total_streaming_time.get, reverse=True)[:10]

# Get the total streaming time for each artist's most streamed track
top_artists_most_streamed = {artist: streaming_history[(streaming_history['artistName'] == artist) & (streaming_history['trackName'] == most_streamed_track['trackName'])]['timeStreaming'].sum().total_seconds() for artist in top_artists}

# Plotting
plt.figure(figsize=(10, 10))
plt.scatter([top_artists_most_streamed[artist] / 3600 for artist in top_artists], [artist_total_streaming_time[artist].total_seconds() / 3600 for artist in top_artists])
plt.xlabel('Total Streaming Time of Most Streamed Track (hours)')
plt.ylabel('Total Streaming Time of All Tracks by Artist (hours)')
plt.title('Correlation Between Most Streamed Track and Artist')
for artist in top_artists:
    plt.text(top_artists_most_streamed[artist] / 3600, artist_total_streaming_time[artist].total_seconds() / 3600, artist)
plt.show()
```
